```python
import jsonlines
import matplotlib as mpl
import matplotlib.pyplot as plt

legislatur = 20

alleReden = []
with jsonlines.open(f'../../data/speeches_{legislatur}.jsonl') as f:
    for line in f.iter():
        #for line in list(f):
        alleReden.append(line)

alleReden.sort(key = lambda x:x['date'])
```

```python
input_list = []
wort_haeufigkeiten =[]

for rede in alleReden:
    #if len(rede['text_lem'])>0:
    if len(rede['text'])>0:
        #if rede['name']=='Angela Merkel':
        #wort_haeufigkeiten.append(len(rede['text_lem']))
        wort_haeufigkeiten.append(len(rede['text']))
        input_dict = {
            'name'  : rede["name"],
            'party' : rede["party"],
        }
        input_list.append(input_dict)
```

```python
mpl.rcParams["figure.figsize"] = (8,4)

plt.hist(wort_haeufigkeiten, bins ='auto')
plt.title("Verteilung der Häufigkeiten")
plt.xlabel("Länge einer Rede (in Zeichen)")
plt.ylabel("Anzahl der Reden mit x Zeichen")
plt.xscale('log')
plt.show()
```

```python

selectionZeichen = []
dropoutZeichen = []
for rede in alleReden:
    #if len(rede['text_lem'])>0:
    if len(rede['text'])>1500:
        #if rede['name']=='Angela Merkel':
        #wort_haeufigkeiten.append(len(rede['text_lem']))
        selectionZeichen.append(rede)
    else:
        dropoutZeichen.append(rede)

print(len(selectionZeichen),len(dropoutZeichen))
```
